```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as nm 
import seaborn as sn
```

```python
path = 'https://raw.githubusercontent.com/ovibaridar/Data_sets/main/weight-height.csv'
```

```python
data = pd.read_csv(path)
```

```python
data.head(3)
```

```python
sn.histplot(data['Height'] , kde = True)
```

```python
sn.histplot(data['Weight'] , kde = True)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
```

```python
data["StandardScaler"] = std.fit_transform(data[["Height"]])
```

```python
data.head(3)
```

```python

data = data[(data["StandardScaler"] < 3) & (data["StandardScaler"] > -3) ]
```

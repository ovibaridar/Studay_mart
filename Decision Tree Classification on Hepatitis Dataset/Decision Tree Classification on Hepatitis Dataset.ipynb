# Remove Null Values from Data 

```python
for col in data.columns:
        null_count = data[col].isnull().sum()
        if null_count > 0:
            data[col].fillna(data[col].mean(), inplace=True)

        
```

# Split the dataset into training (70%) and testing sets.

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state=40 , test_size=.25)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
sns.heatmap(xtrain.corr(), annot=True)
```

#   - Build a Decision Tree classifier using a suitable library (scikit-learn).


```python
from sklearn.tree import DecisionTreeClassifier
import warnings
warnings.filterwarnings("ignore")
```

```python
dtc = DecisionTreeClassifier()
```

#   - Train the model on the training set.

```python
dtc.fit(xtrain,ytrain)
```

```python
dtc.score(xtest,ytest)
```

```python
dtc.score(xtrain,ytrain)
```

#   - Make predictions on the testing set.

```python
perd= dtc.predict(xtest)
perd
```

```python
from sklearn.metrics import confusion_matrix ,precision_score ,recall_score ,accuracy_score , f1_score , RocCurveDisplay
```

```python
cm = confusion_matrix(ytest, perd)
cm
```

```python
sns.heatmap(cm , annot=True)
```

```python
precision_score(ytest,perd)
```

```python
recall_score(ytest,perd)
```

```python
accuracy_score(ytest,perd)
```

```python
f1_score(ytest,perd)
```

```python
from matplotlib import pyplot as plt
RocCurveDisplay.from_predictions(ytest,perd)
plt.plot([0,1],[0,1])
```
